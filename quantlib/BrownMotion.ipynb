```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define dt, sqrt_dt
T = 1
N = 10000
dt = T / N
sqrt_dt = math.sqrt(dt)
```

```python
# Generate Series of 1) time, 2) white noises, 3) Brown motion
t = np.array([i for i in np.arange(0, T, T/N)])
dX = np.random.randn(N) * sqrt_dt
X = np.cumsum(dX)
```

```python
# print graph
plt.subplot(2,1,1)
plt.plot(t, dX)
plt.subplot(2,1,2)
plt.plot(t, X)
```
