```python
from scipy.fft import fft, fftfreq
import numpy as np
import matplotlib.pyplot as plt
```

time vector

```python
time_step = 0.001
T = 1
t = np.arange(0,T, time_step)
N = len(t)
```

synthetic data

```python
freq1 = 10
freq2 = 25
y = 1*np.sin(2.0*np.pi*freq1*t) + 0.5*np.sin(2.0*np.pi*freq2*t)
```

plot the synthetic data

```python
plt.plot(t,y)
plt.ylabel('Signal')
plt.xlabel('Time')
plt.grid()
plt.show()
```

fft

```python
yhat= fft(y)
yhat = yhat[0:N//2]
yhat_amplitude = 2.0 *np.abs(yhat)*time_step
```

frequency vector

```python
f = fftfreq(N, time_step)
f = f[0:N//2]
```

fft plot

```python
plt.plot(f, yhat_amplitude, 'o-')
plt.ylabel('Amplitude')
plt.ylabel('frequency')
plt.xlim([0,50])
plt.grid()
plt.show()
```
