```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
y = 2*x
```

```python
print ('x=', x)
print ('y=', y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, 'b-o', label='y=2x')
plt.ylabel('Y data')
plt.xlabel('X data')
plt.legend()
plt.grid()
```

```python
fig, ax = plt.subplots()
ax.plot=(x, y, 'b-o')
ax.set_ylabel('Y data')
ax.set_xlabel('x data')
ax.legend(['y=2x'])
ax.grid()
```

```python
x = np.linspace(-5, 5, 11)
y = x**2
plt.plot(x,y, 'b-o', label='$y=x^{2}$')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.grid()
```

```python
x = np.arange(0, 2*np.pi, 0.2)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1, 'b-o', label='y=xin(x)')
plt.plot(x,y2, 'b--', label='y=cos(x)')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.grid()
```
