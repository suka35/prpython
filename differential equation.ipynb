```python
import numpy as np
from scipy.integrate import ode
import matplotlib.pyplot as plt
 
 
def odefun(t, Z, omega):
    dZdt = np.zeros(2)
    dZdt[0] = Z[1]
    dZdt[1] = - omega**2 * Z[0]
    return dZdt
 
 
A = 1
omega = 2
t0, tf = 0, 10
Z0 = np.array([A, 0])
 
solver = ode(odefun).set_integrator('dopri5')
 
sol = []
 
 
def solout(t, Z):
    sol.append([t, *Z])
 
 
solver.set_solout(solout)
solver.set_initial_value(Z0, t0).set_f_params(omega)
solver.integrate(tf)
 
sol = np.array(sol)
 
plt.plot(sol[:, 0], sol[:, 1])
plt.plot(sol[:, 0], sol[:, 2])
plt.legend(['position', 'velocity'])
plt.grid(True, linestyle='--')
```
